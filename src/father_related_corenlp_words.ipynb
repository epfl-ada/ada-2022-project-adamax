```python
import pandas as pd
import matplotlib.pyplot as plt
from helpers import *
import nltk
from wordcloud import WordCloud
# nltk.download()
```

```python
# Import father related words
dep, gov = get_father_words_from_json()
```

```python
# Read the dicts to data frames with columns word and count
dep_df = pd.DataFrame.from_dict(dep, orient='index', columns=['count'])
gov_df = pd.DataFrame.from_dict(gov, orient='index', columns=['count'])

# Sort the data frames by count
dep_df = dep_df.sort_values(by=['count'], ascending=False)
gov_df = gov_df.sort_values(by=['count'], ascending=False)

# Plot the top 20 words
dep_df[:20].plot(kind='bar')
plt.show()
gov_df[:20].plot(kind='bar')
plt.show()
```

```python
# Use nltk and pos_tag to get the pos (part of speech) of the words
dep_df['pos'] = dep_df.index.map(lambda x: nltk.pos_tag([x])[0][1])
gov_df['pos'] = gov_df.index.map(lambda x: nltk.pos_tag([x])[0][1])

# Filter the data frames to only keep the adjectives
dep_df_adj = dep_df[dep_df['pos'] == 'JJ']
gov_df_adj = gov_df[gov_df['pos'] == 'JJ']

# Plot the data
dep_df_adj.head(20).plot.bar()
title = 'Top 20 adjectives that depend on father'
plt.title(title)
plt.show()

gov_df_adj.head(20).plot.bar()
title = 'Top 20 adjectives that are governed by father'
plt.title(title)
plt.show()
```

```python
# Plot word clouds of the father related adjectives

# Create a word cloud of the adjectives that depend on father
dep_wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(dep_df_adj['count'].head(50))
plt.figure(figsize=(10, 5))
plt.imshow(dep_wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()

# Create a word cloud of the adjectives that are governed by father
gov_wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(gov_df_adj['count'].head(50))
plt.figure(figsize=(10, 5))
plt.imshow(gov_wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
